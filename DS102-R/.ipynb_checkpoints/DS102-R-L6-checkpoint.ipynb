## Requirements

From the `mtcars` data frame, create a box plot of miles per gallon (the `mpg` variable) grouped by the number of cylinders in the engine (the `cyl` variable). Do these box plots make sense? Also, use the `summarize()` and `group_by()` functions to compute how many cars have four cylinders, how many have six, and how many have eight.

# Install and Import Packages

```r
library(ggplot2)
library(dplyr)
```

# Look at the Dataset

```r
head(mtcars)
```

# Create a box plot of miles per gallon (the `mpg` variable) and number of cylinders (the `cyl` variable)

```r
# My work
boxplot(mpg~cyl,data=mtcars, main="Car Milage Data", xlab="Number of Cylinders", ylab="Miles Per Gallon")
```

```r
# The solution
boxplot(mtcars$mpg ~ mtcars$cyl, data=mtcars)
```

```r
# The slightly more complex way with more adaptability, using ggplot
ggplot(mtcars, aes(x = cyl, y = mpg)) + geom_boxplot(aes(group=cyl))
```

```r
# Use the summarize and group by function to compute how many cars have 
# each cylinder number
mtcars %>% group_by(cyl) %>% summarize(count = n())
```

# Conclusion:
### Yes, this makes sense - the more cylinders a car has, the more gas it will use. The lower the miles per gallon less cylinders the car will have.
