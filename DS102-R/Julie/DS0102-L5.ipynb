#
# Import Packages

```r
library (ggplot2)
library (tidyverse)
```

#
# Look at the Dataset and change the Built in Dataset "rivers" to a dataframe

```r
rr <- data.frame(rivers)
```

```r
head(rivers)
```

#
# Histogram Plot

```r
rr1_histo <- ggplot(rr,aes(x= rivers))
rr1_histo + geom_histogram(binwidth= 100) + ggtitle("Major Rivers in North America")
```

#
# Box Plot

```r
rr1_box <- ggplot (rr, aes(x= "", y= rivers)) 
rr1_box + geom_boxplot () + xlab ("")
```

#
# FINDING OUTLINERS ON BOXPLOT 
### Summary function tells you the quartiles

```r
summary(rivers)
```

#
# Finding Outliners 
### 3rd Qrt - 1st Qrt 

```r
IQR <- 680.0 - 310.0
```

```r
outliner_range <- IQR * 1.5
```

```r
680.0 + outliner_range
```

```r
310.0 - outliner_range
```

```r
# Anything below -245 and above 1235 are the outliners.
```

#
# Normal Distribution Plot

```r
ggplot (rr, aes(sample= rivers)) + geom_qq()
```

```r
# You can see the points around the 1235 & -245 start becoming outliners
# This data is most likely accurate and all the data is lining up 
```
